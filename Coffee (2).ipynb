```python
missing_color_df1 = df_q[df_q['Variety'].isna()][['Country.of.Origin']].value_counts()
print(missing_color_df1)
```

```python
missing_color_df2 = df_q[df_q['Variety'].notna()][['Country.of.Origin']].value_counts()
print(missing_color_df2)
```

```python

df_q['Variety'] = df_q.groupby('Country.of.Origin')['Variety'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df_q.isnull().sum()
#mapping of every country and its corresponding variety using forward fill and backward fill
```

```python
missing_color_df1 = df_q[df_q['Variety'].isna()][['Country.of.Origin']].value_counts()
print(missing_color_df1)
#these don't have a corresponding in any row so it fills with mode
```

```python
df_q['Variety'].fillna(df_q['Variety'].mode()[0],inplace=True)
df_q['Variety'].value_counts().plot.bar()
```
